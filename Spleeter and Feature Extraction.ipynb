## Feature Extraction

### Accompaniment

```python
import numpy as np

header = header = 'filename chroma_stft rms spectral_centroid spectral_bandwidth rolloff zero_crossing_rate'
for i in range(1, 21):
    header += f' mfcc{i}'
header += ' label'
header = header.split()
```

```python
rows = []
i = 0

for folder_name in os.listdir('../Spleeter_Excerpt_Output/'):
    
    input_path = '../Spleeter_Excerpt_Output/{}/accompaniment.wav'.format(folder_name)
    y, sr = librosa.core.load(input_path, mono=True, sr = 44100)
    
    #Feature Extraction
    chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
    rms = librosa.feature.rms(y=y)
    spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
    spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
    rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
    zcr = librosa.feature.zero_crossing_rate(y)
    mfcc = librosa.feature.mfcc(y = y, sr = sr)
    
    to_append = [folder_name, np.mean(chroma_stft),np.mean(rms),np.mean(spec_cent),np.mean(spec_bw), np.mean(rolloff), np.mean(zcr)]    
    for e in mfcc:
        to_append.append(np.mean(e))
    to_append.append(int(folder_name[5]))
    rows.append(to_append)  #Append the row list to a 2D list called 'rows'
    if i%20 == 0 or i<=10:
        print(folder_name, " completed")
    i += 1
```

```python
import csv
file = open('../../Data_Accompaniment_Only_30s_Overlapping.csv', 'w', newline= '')
with file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(rows)
```
